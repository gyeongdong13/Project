```python
deposit = pd.read_csv('/content/drive/MyDrive/데이터 청년 캠퍼스/농협 프로젝트/데이터 모음/변수선택(예금).csv', encoding='utf-8')

deposit.head()
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
df = sc.fit_transform(deposit.drop(columns=["사무소명", "2022 목표", "2022 실적", "2021 목표", "2021 실적", "2020 목표", "2020 실적"]))
columns=["22년 평잔내역", "21년 평잔내역","20년 평잔내역", "22년 영업이익", "21년 영업이익", "20년 영업이익", "2022 달성률", "2021 달성률", "2020 달성률", "2022 직원", "2021 직원", "2020 직원", '2022 생명보험 건수', '2021 생명보험 건수', '2020 생명보험 건수', '2022 생명보험 수수료', '2021 생명보험 수수료', '2020 생명보험 수수료', "20년 단말기", "21년 단말기", "22년 단말기"]
df = pd.DataFrame(df, columns=columns)
df.head(5)
```

# 20년 직원 추출현황

```python
# 종속변수 지정(2020년)
y = deposit["2020 직원"]
y.head(5)
```

```python
# 독립변수 지정
X = df[['20년 평잔내역', '20년 영업이익', '2020 달성률', '2020 생명보험 건수', '2020 생명보험 수수료', '20년 단말기']]
X.head(5)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import Lasso
X_train_lasso, X_val_lasso, y_train_lasso, y_val_lasso = train_test_split(X, y, test_size=0.1, random_state=0)

max_score = 0
max_alpha = 0
alphas = [10 ** i for i in range(-1, 6)]
for alpha in alphas: # alpha를 0.1부터 10000까지 10배씩 증가
    lasso_clf = Lasso(alpha=alpha) #라쏘 분류기 선언
    lasso_clf.fit(X_train_lasso, y_train_lasso) #학습
    score = lasso_clf.score(X_val_lasso, y_val_lasso) #검정 데이터셋에 대해 정확도를 측정
    if score > max_score: #정확도가 이전의 정확도 최대값보다 크면 최대값을 변경
        max_score = score
        max_alpha = alpha
print('Max alpha {:.3f} at max validation score {:.3f}'.format(max_alpha, max_score))
```

```python
from sklearn.linear_model import Lasso
import numpy as np

lasso_clf = Lasso(alpha=0.1)
lasso_clf.fit(X_train, y_train)

print('#Used features count: {}'.format(np.sum(lasso_clf.coef_ != 0)), 'out of', X.shape[1])
```

```python
# 회귀계수가 0이 아닌 변수들의 컬럼 이름 확인
selected_feature_names = X.columns[lasso_clf.coef_ != 0]
print("Selected feature names:", selected_feature_names.tolist())
```

```python
X = df[['20년 평잔내역', '20년 영업이익', '2020 달성률', '2020 생명보험 수수료', '20년 단말기']]
y = deposit["2020 직원"]

import statsmodels.api as sm
# 상수항 추가
X = sm.add_constant(X)

# 선형 회귀 모델 적합
linear_model = sm.OLS(y, X)
linear_results = linear_model.fit()

# 결과 출력
print(linear_results.summary())
```

```python
import statsmodels.api as sm

# 독립 변수
X = df[['20년 평잔내역', '20년 영업이익', '2020 달성률', '2020 생명보험 수수료']]
# 종속 변수 (빈도 데이터)
y = deposit["2020 직원"]

# 상수항 추가
X = sm.add_constant(X)

# 포아송 회귀 모델 적합
poisson_model = sm.GLM(y, X, family=sm.families.Poisson())
poisson_results = poisson_model.fit()

# 결과 출력
print(poisson_results.summary())
```
